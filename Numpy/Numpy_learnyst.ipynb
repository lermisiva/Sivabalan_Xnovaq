```python
# 3. Indexing & Slicing
# #Given a 2D array of shape (10, 10), use advanced integer indexing to extract elements at
# random row and column indices, including duplicate indices.
# # Extract a sub-array excluding the border rows and columns (i.e., inner (8, 8) block).
# #Reverse the rows of a matrix without changing the columns using slicing.
arr = np.arange(100).reshape(10, 10)
arr1 = np.random.randint(1,10 ,size= 5)
arr2 = np.random.randint(1,10 ,size= 5)

print("row :\n", arr1,"\n")
print("column :\n", arr2,"\n")

mer = arr[arr1, arr2]
print(mer)

inner = arr[1:-1, 1:-1]
print("Inner 8x8 block:\n", inner)

rev_rows = arr[::-1, :]
print("Rows reversed and columns unchanged:\n", rev_rows)
```

```python
#4. Reshaping & Manipulating Arrays
## Reshape a (100,) array into (10, 10) and then flatten it back, verifying the integrity of the
#  data.
##Concatenate three arrays along different axes; demonstrate horizontal and vertical
# stacking with appropriate code.
## Split a (12, 12) matrix vertically into 3 equal parts using NumPy splitting functions.

# Create array of shape (100,)
arr = np.arange(100)
print("Original Shape:", arr.shape)

# Reshape to (10, 10)
reshaped = arr.reshape(10, 10)
print("Reshaped (10x10):\n", reshaped)

# Flatten back to 1D
flattened = reshaped.flatten()
print("Flattened Shape:", flattened.shape)

# Integrity check
print("Integrity Check (same data):", np.array_equal(arr, flattened))


# Create 3 arrays
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
c = np.array([[9,10],[11,12]])

print("Array A:\n", a)
print("Array B:\n", b)
print("Array C:\n", c)

# Horizontal stacking
h_stack = np.hstack((a, b, c))
print("\nHorizontal Stack:\n", h_stack)

# Vertical stacking
v_stack = np.vstack((a, b, c))
print("\nVertical Stack:\n", v_stack)

# Concatenate axis=0
con0 = np.concatenate((a, b, c), axis=0)
print("\nConcatenate axis=0:\n", con0)

# Concatenate axis=1
con1 = np.concatenate((a, b, c), axis=1)
print("\nConcatenate axis=1:\n", con1)

# Create 12x12 matrix
matrix = np.arange(144).reshape(12, 12)
print("Original 12x12 Matrix:\n", matrix)

# Split vertically (columns)
vertical_parts = np.split(matrix, 3, axis=1)

print("\nVertical Part 1:\n", vertical_parts[0])
print("\nVertical Part 2:\n", vertical_parts[1])
print("\nVertical Part 3:\n", vertical_parts[2])
```

```python
# 5. Aggregation Functions
## Calculate the mean, median, and standard deviation of each column in a (15, 10) array. 
##Find the indices of minimum and maximum values along each row.
##Compute the cumulative sum and cumulative product of elements along the flattened
# array.

arr = np.random.randint(1, 100, size=(15, 10))
print("Array (15x10):\n", arr)

# Column wise
mean_cols = np.mean(arr, axis=0)
median_cols = np.median(arr, axis=0)
std_cols = np.std(arr, axis=0)

print("\nColumn-wise Mean:\n", mean_cols)
print("\nColumn-wise Median:\n", median_cols)
print("\nColumn-wise Std Dev:\n", std_cols)


min_indices = np.argmin(arr, axis=1)
max_indices = np.argmax(arr, axis=1)

print("\nIndices of Minimum Value in Each Row:\n", min_indices)
print("\nIndices of Maximum Value in Each Row:\n", max_indices)

flat = arr.flatten()

cum_sum = np.cumsum(flat)
cum_prod = np.cumprod(flat)

print("\nCumulative Sum of Flattened Array:\n", cum_sum)
print("\nCumulative Product of Flattened Array:\n", cum_prod)
```

```python
# 6. Handling Missing Values
# #Create an array of 200 random floats and randomly assign 10% of values to np.nan. 
# #Write code to detect and count the missing values. Replace missing values with the mean of their respective columns.
# #Demonstrate removing rows or columns containing any missing values.

arr = np.random.random(200)


num_missing = int(0.10 * arr.size)     
missing_indices = np.random.choice(arr.size, num_missing, replace=False)

arr[missing_indices] = np.nan
print("Array with NaN values:\n", arr)


arr = arr.reshape(20, 10)
print("\nReshaped (20x10):\n", arr)

missing_mask = np.isnan(arr)

missing_count = np.sum(missing_mask)

print("\nMissing Value Mask:\n", missing_mask)
print("\nTotal Missing Values:", missing_count)

col_means = np.nanmean(arr, axis=0)


arr_filled = np.where(np.isnan(arr), col_means, arr)

print("\nColumn Means:\n", col_means)
print("\nArray After Replacing NaN with Column Means:\n", arr_filled)


rows_removed = arr[~np.isnan(arr).any(axis=1)]
print("\nRows with no missing values:\n", rows_removed)

cols_removed = arr[:, ~np.isnan(arr).any(axis=0)]
print("\nColumns with no missing values:\n", cols_removed)
```
