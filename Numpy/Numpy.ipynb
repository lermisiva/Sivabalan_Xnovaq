```python
# 1.Create a 2D NumPy array (10×5) with random integers (1–100).
import numpy as np

arr = np.random.randint(1, 101, size=(10, 5)) # size = (num of rows, num of columns)
print(arr)
```

```python
# 2. Replace all values < 50 with 0 and ≥ 50 with 1.
rep = np.where(arr<50,0, 1) # np.where(condition, value_if_true, value_if_false)
print(rep)
```

```python
# 3. Compute mean, standard deviation, variance along rows and columns.
print("---------Rows--------")
row_mean = np.mean(arr, axis = 1)
print("Mean Row = ",row_mean)
row_std = np.std(arr, axis = 1)
print("standard Deviation Row = ",row_std)
row_var = np.var(arr, axis = 1)
print("Variance Row = ",row_var)

# column 
print("------------Column----------")
Column_mean = np.mean(arr, axis = 0)
print("Mean Row = ",Column_mean)
Column_std = np.std(arr, axis = 0)
print("standard Deviation Row = ",Column_std)
Column_var = np.var(arr, axis = 0)
print("Variance Row = ",Column_var)
```

```python
# 4. Generate two arrays A(3×3) and B(3×3) and perform matrix multiplication.
a = np.random.randint(1,10, size =( 3,3))
print("Value of A :\n",a)
b = np.random.randint(1,10, size =( 3,3))
print("Value of B :\n",b)

c = np.matmul(a,b)
print("Mul of A, B :\n",c)
```
