# Mean

```python
a =[50,51,53,59.5,14,16,20,21,29,52,54,60,67,90,53,52,50]
n = len(a)
sum = 0
for i in a :
    sum +=i
# print(sum)
m = sum/n
print(m)
```

# Median

```python
a =[50,51,53,59.5,14,16,20,21,29,52,54,60,67,90,53,52,50]
b = sorted(a)
# print(b)
n = len(a)
if n%2 != 0:
    m = int((n+1)/2)
    print(b[m])
else:
    x =b[int(n/2)]
    y =b[int((n+1)/2)]
    z = (x+y)/2
    print(z)
    
    
```
