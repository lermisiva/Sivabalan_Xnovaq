```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import os
```

```python
os.chdir ('C:\\Noble\\Training\\KODO\\2025\\Dec 2025\\Dec 2025 Day 13\\')
```

```python
df = pd.read_csv ('mtcars_for_manymerge.csv')
print (df.shape)
print (df)
```

```python
x = df.iloc[:,1:-1]
print (x)
```

```python
y = df['Feedback']
print (y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y) # Default test size 25 %
print (x_train.shape)
print (x_test.shape)
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
metrics.accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
knn = KNeighborsClassifier(n_neighbors=4)
knn_accuracies = cross_val_score(knn, x, y, cv=5)
print (pd.DataFrame(knn_accuracies))  # Individual Accuracies 
```

```python
print ('Mean', knn_accuracies.mean())
print ('STD',knn_accuracies.std())
```

```python

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logi_accuracies = cross_val_score(logreg, x, y, cv=5)
print (pd.DataFrame(logi_accuracies))
```
