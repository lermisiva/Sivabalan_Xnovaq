```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
tips =sns.load_dataset('tips')
tips1= tips
print (tips1)
```

```python
sns.boxplot (data = tips1 , x = 'day', y = 'total_bill' )
plt.show()
```

```python

print (tips [tips1['day'] =='Fri'].describe())
```

```python

sns.displot(tips [tips1['day'] =='Fri']['total_bill'], kde = True)
plt.show()
```

```python
sns.displot(tips1['total_bill'], kde = True)
plt.show()
```
