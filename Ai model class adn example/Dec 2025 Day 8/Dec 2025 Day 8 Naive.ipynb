```python

x= df1.iloc [:,2:4].values
y= df1.iloc[:,4].values
from sklearn.preprocessing import MinMaxScaler
minmax_x= MinMaxScaler()
minmax_x = minmax_x.fit_transform(x)
print (minmax_x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(minmax_x,y,test_size= 0.2)
print (x.shape)
print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python

from sklearn.naive_bayes import MultinomialNB
MNB = MultinomialNB()
MNB.fit(x_train,y_train)
print (MNB)
```

```python
MNB_pred = MNB.predict(minmax_x)
print (MNB_pred)
```

```python
from sklearn.metrics import accuracy_score
print (accuracy_score (y,MNB_pred))
```

```python

from sklearn.metrics import classification_report
print(classification_report (y,MNB_pred,zero_division=0))
```

```python
cm = confusion_matrix (y, MNB_pred)
print (cm)	
```
