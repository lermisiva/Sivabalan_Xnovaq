```python
data = pd.read_csv('pacific.csv')
display (data)
```

```python

print (data.groupby('Status').agg({'Status': np.size}))
```

```python
data.dtypes
```

```python
data.Status = pd.Categorical(data.Status)
```

```python
data.dtypes
```

```python

data['Status'] = data.Status.cat.codes
display (data.head())
```

```python

print (data.groupby('Status').agg({'Status': np.size}))
```

```python
pred_columns = data[:]
pred_columns.drop(['Status'],axis=1,inplace=True)
pred_columns.drop(['Event'],axis=1,inplace=True)
pred_columns.drop(['Latitude'],axis=1,inplace=True)
pred_columns.drop(['Longitude'],axis=1,inplace=True)
pred_columns.drop(['ID'],axis=1,inplace=True)
pred_columns.drop(['Name'],axis=1,inplace=True)
pred_columns.drop(['Date'],axis=1,inplace=True)
pred_columns.drop(['Time'],axis=1,inplace=True)
display((pred_columns))
```

```python
x = pred_columns.values
display (x)
```

```python
y =data['Status'].values
print (y)
```

```python
from sklearn.model_selection  import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
print (x.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
log_pred = logmodel.predict(x)
print (log_pred)
print (pd.DataFrame(log_pred).groupby(0).agg({0: np.size}))
print('\n Accuracy Score')
print (accuracy_score (y,log_pred))
print('\nClassification Report')
print(classification_report (y,log_pred))
print('Confusion Matrix')
print (confusion_matrix (y,log_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN1 = KNeighborsClassifier(n_neighbors =100 , metric = 'minkowski', p= 2)
KNN1.fit(x_train,y_train)
KNN1_pred = KNN1.predict(x)
print (KNN1_pred)
print (pd.DataFrame(KNN1_pred).groupby(0).agg({0: np.size}))
print('\n Accuracy Score')
print (accuracy_score (y,KNN1_pred))
print('\nClassification Report')
print(classification_report (y,KNN1_pred))
print('Confusion Matrix')
print (confusion_matrix (y,KNN1_pred))

```
