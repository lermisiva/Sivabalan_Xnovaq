```python

vector = CountVectorizer()
vec = vector.fit_transform(df.text)
bow= vec.toarray()
print (bow.shape)
display (bow)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
label = df.spam
classifier.fit(vec, label)
```

```python
pred = classifier.predict (vec)
print (pred.shape)
print (pred)
```

```python
from sklearn.metrics import accuracy_score
print (accuracy_score (label,pred))
```

```python

from sklearn.metrics import classification_report
print(classification_report(label,pred,zero_division=0))

```

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix (label,pred)
print (cm)	
```

```python
df_pred = pd.DataFrame (pred,columns = ['Spam Prediction']) 
result = pd.concat ([df,df_pred ],axis = 1)
result.to_csv('08_Jan_Spam_Prediction.csv')
display (result)
```
