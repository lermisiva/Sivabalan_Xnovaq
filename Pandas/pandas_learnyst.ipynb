```python
# 1. Load the dataset in Pandas
import pandas as pd
student = pd.read_excel("C:\\Users\\XNovaQ.23\\Downloads\\students_assignment_data.xlsx")
print(student)
```

```python
#2. Display first 15 rows, data types, and dataset shape.
h = student.head(15)
print( "Data set shape:",h.shape)
print("Data Types :\n",h.dtypes)
```

```python
#3. Identify unique cities and count of students per city.
city = student['City'].nunique()
print("Unique_City Count :",city)
```

```python
# 4. Filter and display students with Total_Marks greater than or equal to 400.
t = student[student['Total_Marks'] >= 400]
print(t)
```

```python
# 5. Extract details of all female students and their scores.
fem = student[student['Gender'] == 'F'][['Name','Total_Marks']]
print(fem)
```

```python
# 6 .Group students by City and compute the average Total_Marks and the total count per city.
g = student.groupby('City').agg(
    avg_marks=('Total_Marks', 'mean'),
    Count=('StudentID', 'count')
)

print(g)
    
```

```python
# 7 .Sort all students by Total_Marks in descending order.
sorted_students = student.sort_values(by='Total_Marks', ascending=False)
print(sorted_students)
```

```python
#8 . Reset the index of the sorted DataFrame.
sorted_students = sorted_students.reset_index(drop=True)

print(sorted_students)
```

```python
#9 . Detect missing Total_Marks values (simulate by setting ~10% missing).
import pandas as pd
import numpy as np

missing_indices = student.sample(frac=0.10, random_state=42).index
student.loc[missing_indices, 'Total_Marks'] = np.nan
missing_mask = student['Total_Marks'].isna()
missing_count = student['Total_Marks'].isna().sum()
print("\nTotal Missing Values in Total_Marks:", missing_count)
print("\nRows with Missing Total_Marks:")
print(student[student['Total_Marks'].isna()])
```

```python
#10. Impute missing Total_Marks with the mean of the column.
mean_Value = student['Total_Marks'].mean()
fill =student['Total_Marks'].fillna(mean_Value, inplace =True)
print(student)
```

```python
# 11. Create a new column Result where students scoring >= 200 are marked as 'Pass', otherwise 'Fail'.
student['Result'] = student['Total_Marks'].apply((lambda x : 'Pass' if x>=200 else 'Fail'))
student
```

```python
# 12. Count how many students failed and passed overall.
student['Result'].value_counts()
```
